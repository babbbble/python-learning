### Part 2: Fetching sports scores from an API
Now, you'll write a function that fetches the latest sports scores using an API. We'll use the Football Data API (https://www.football-data.org/) to get current soccer match results and/or live scores. To do this, you will first need to get an API Key by heading over to the Football Data API, signing up for a free account, and grabbing your API key. Then you can use the script below to fetch sports scores.

```python
# Get your API key from https://www.football-data.org/ and replace 'YOUR_API_KEY_HERE'
api_key = 'a66eddf96f214331a01074196fe9e754'
headers = {'X-Auth-Token': api_key}
url = "http://api.football-data.org/v4/matches"
def fetch_scores():
    response = requests.get(url, headers=headers)
    try:
        data = response.json()
# Check if the 'matches' key exists in the response
        if 'matches' in data and len(data['matches']) > 0:
            match = data['matches'][0]
            home_score = match['score']['fullTime'].get('home')
            away_score = match['score']['fullTime'].get('away')
            score_line = f"{home_score if home_score is not None else 'N/A'} - {away_score if away_score is not None else 'N/A'}"
            team_home = match['homeTeam']['name']
            team_away = match['awayTeam']['name']
            return f"{team_home} {score_line} {team_away}"
        else:
            return "No matches currently available."
    except Exception as e:
        return f"Error fetching data: {e}"
fetch_scores()
```

### Part 3: Creating notifications
Once you have the sports data, you can send desktop notifications using the `plyer` library. Your code should result in a notification with the latest soccer statistics.

```python
import subprocess

def send_notification():
    match_score = fetch_scores()
    safe_message = str(match_score).replace('"', '\\"')

    subprocess.run([
        "osascript",
        "-e",
        f'display notification "{safe_message}" with title "Live Sports Score Update"'
    ])


# Test sending a notification
send_notification()
```

### Part 4: Automating notifications
Now that you have successfully retrieved scores from an API and prompted notifications, let's automate this process to run regularly. When scheduling automated processes, modify the interval or time of day the notifications are sent. There are numerous other conditions you might want to check before displaying a notification, such as the time of day. 

Remember that you want to schedule the sports data to be fetched and sent to your desktop every morning at 8:00am. Most of this code is shown below, except for a portion of the first line. Fill in the first line of the code below, then use it in your notebook to schedule this process.

```python
# Schedule notifications every day at 08:00
schedule.every().day.at("08:00").do(send_notification)

# Test the scheduling of notifications (this will run until you stop it)
while True:
    schedule.run_pending()
    time.sleep(1)
```
